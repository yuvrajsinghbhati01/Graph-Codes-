```python
# Q1. Write a code to find the degree of each vertex, and store it in a dictionary. Sort the keys of the dictionary by the degree stored in the values.

def nodes_degree(given_adjacency_list) :
    # This for loop will iterate over each key of adjacency list and find its degree and store it in dictionary
    degree_of_nodes = {}
    for key in given_adjacency_list.keys() :
        degree_of_nodes[key] = len(given_adjacency_list[key])

    keys_list = list(degree_of_nodes.keys())

    # These for loops will iterate over elements of keys list and check if value of keys of degree of nodes dictionary is sorted or not
    for i in range(len(keys_list)) :
        for j in range(i+1,len(keys_list)) :
            if degree_of_nodes[keys_list[j]] > degree_of_nodes[keys_list[i]] :
                keys_list[i] , keys_list[j] = keys_list[j] , keys_list[i] 
    return keys_list
```

```python
# Q2. Write a code to inter-convert the 3 graph representations we have learnt.

def adjacencyList_to_adjacencyMatrix(given_adjacency_list , nodes_list) :
    adjacency_matrix = []

    # Outer loop will iterate over keys of my dict dictionary and inner loop will iterate over nodes names and check if ith node is in adjacent node list then append 1 otherwise 0 and after every key append that list in matrix
    for key in given_adjacency_list.keys() :
        row = []
        for i in range(len(nodes_list)) :
            if nodes_list[i] in given_adjacency_list[key] :
                row.append(1)
            else :
                row.append(0)
        adjacency_matrix.append(row)
    return adjacency_matrix


def adjacencyMatrix_to_edgeList(given_adjacency_matrix , nodes_names) :
    edge_list = []

    # Both loop iterate over the nodes name list and check if ith and jth node of nodes name list are adjacent to each other then add this edge in edge list 
    for i in range(len(nodes_names)) :
        for j in range(len(nodes_names)) :
            if given_adjacency_matrix[i][j] == 1 :
                edge_list.append((nodes_names[i],nodes_names[j]))
    return edge_list 



def edgeList_to_adjacencyList(given_edge_list , nodes_names) :
    adjacency_list = {}

    # It will iterate over all the tuples of my list and check if 1st element of tuple is present in keys of adjacency list dictionary then it will append 2nd element of that tuple in value of key of adjacency list dictionary otherwise it will add 1st element of tuple as key and 2nd element as value in adjacent list dictionary
    for i in range(len(given_edge_list)) :
        if given_edge_list[i][0] in adjacency_list.keys() :
            adjacency_list[given_edge_list[i][0]].append(given_edge_list[i][1])
        else :
            adjacency_list[given_edge_list[i][0]] = []
            adjacency_list[given_edge_list[i][0]].append(given_edge_list[i][1])
    return adjacency_list

```

```python
# Q4. Check if a graph is complete.

def check_using_adjacency_matrix(given_adjacency_matrix) :
    
    # Outer and inner loop will iterate over all rows and columns of adjacency matrix respectively and check if all non diagonal element is 1, i.e, ith and jth element of nodes names have edge between then or they are adjacent to each other
    for i in range(len(given_adjacency_matrix)) :
        for j in range(len(given_adjacency_matrix)) :
            if i == j :
                continue
            elif given_adjacency_matrix[i][j] != 1 :
                return False 
    return True
```

```python
# Q5. Check if a graph is connected.

def using_adjacency_list(my_list, nodes_names) :
    visited_list = []
    queue = []
    front = rear = -1 

    for key in my_list.keys() :
        adjacent_nodes = my_list[key]      

        # It will check if node is not present in visited list then it will add both in visited list and queue     
        if key not in visited_list :
            if len(visited_list) == 0 : 
                visited_list.append(key)
                queue.append(key)
                front += 1
                rear += 1
            else :
                visited_list.append(key)
                queue.append(key)
                rear += 1

        # It will check if adjacent nodes of the current key is not present in visited list then it will add both in visited list and queue
        for i in range(len(adjacent_nodes)) :
            if adjacent_nodes[i] not in visited_list :
                    visited_list.append(adjacent_nodes[i])
                    queue.append(adjacent_nodes[i])
                    rear += 1
        
        # If queue turns empty then it will check if visited list and nodes names list are equal then the graph is connected otherwise it is not
        queue.pop(front)
        if (not queue) :
            if visited_list == nodes_names :
                    return True 
            else :
                    return False 
        
    
    # After iterating on all keys of adjacency list, queue has some element then remove those element and check if visited list = nodes name list then graph is connected    while queue :
    else :
        if visited_list == nodes_names :
                return True 
        else :
                return False 
        
```

```python
# Q6. Given a graph and a list of vertices, check if it forms a walk, or a trail or a path or none of these.

def check_walk_using_edge_list(given_list , graph_list , nodes_names) :
    edge_taken = []
    nodes_visited = []

    # It is iterating over all nodes of given list and checking if ith and i+1th node are adjacent or not and they exits in graph list or not (to check given sequence is connected)
    for i in range(len(given_list) - 1) :
        node = given_list[i]
        edge = (given_list[i] , given_list[i+1])
        if edge in graph_list  :
            edge_taken.append(edge)
            nodes_visited.append(node)
        else : 
            return False
    nodes_visited.append(given_list[-1])            # Last node will be left, so adding it now
        
    # Creating two lists which stores unique edges taken and unique nodes visited
    distinct_edge_taken = list(set(edge_taken))

    distinct_nodes_visited = []
    for i in nodes_visited :
        if i not in distinct_nodes_visited :
            distinct_nodes_visited.append(i)

    # Path is when nodes and edges repeats, so it checks if edges and nodes are repeated (if repeated then length will be different) or not 
    if len(edge_taken) != len(distinct_edge_taken) and len(nodes_visited) != len(distinct_nodes_visited) :
        return True
    else : 
        return False 
    
def check_trail_using_edge_list(given_list , graph_list , nodes_names) :
    edge_taken = []
    nodes_visited = []

    # It is iterating over all nodes of given list and checking if ith and i+1th node are adjacent or not and they exits in graph list or not (to check given sequence is connected)
    for i in range(len(given_list) - 1) :
        node = given_list[i]
        edge = (given_list[i] , given_list[i+1]) 
        if edge in graph_list :
            edge_taken.append(edge)
            nodes_visited.append(node)
        else :
            return False
    nodes_visited.append(given_list[-1])             # Last node will be left, so adding it now
        
    # Creating two lists which stores unique edges taken and unique nodes visited
    distinct_edge_taken = list(set(edge_taken))
    distinct_nodes_visited = list(set(nodes_visited))

    # Trail is when nodes may repeat but edges won't, so it checks if nodes are repeated (if repeated then length will be different) or not and edges are unique or not 
    if len(edge_taken) == len(distinct_edge_taken) and len(nodes_visited) != len(distinct_nodes_visited) :
        return True
    else :
        return False
    

def check_path_using_edge_list(given_list , graph_list , nodes_names) :
    edge_taken = []
    nodes_visited = [] 

    # It is iterating over all nodes of given list and checking if ith and i+1th node are adjacent or not and they exits in graph list or not (to check given sequence is connected)
    for i in range(len(given_list) - 1) :
        node = given_list[i]
        edge = (given_list[i] , given_list[i+1]) 
        if edge in graph_list :
            edge_taken.append(edge)
            nodes_visited.append(node)
        else :
            return False
    nodes_visited.append(given_list[-1])                # Last node will be left, so adding it now
        
    # Creating two lists which stores unique edges taken and unique nodes visited
    distinct_edge_taken = list(set(edge_taken))
    distinct_nodes_visited = list(set(tuple(nodes_visited)))
    
    # Path is when neither node nor edges repeats, so it checks if nodes and edges are repeated (if repeated then length will be different) or not 
    if len(edge_taken) == len(distinct_edge_taken) and len(nodes_visited) == len(distinct_nodes_visited) :
        return True 
    else : 
        return False

def check_sequence(given_list , graph_list , nodes_names) :
    check_walk = check_walk_using_edge_list(given_list , graph_list , nodes_names)
    check_trail = check_trail_using_edge_list(given_list , graph_list , nodes_names)
    check_path = check_path_using_edge_list(given_list , graph_list , nodes_names)

    if check_walk == True :
        return 'Walk'
    elif check_trail == True :
        return 'Trail'
    elif check_path == True :
        return 'Path'
    else :
        return 'NOTA'
```

```python
# Q7. Check if a given graph is a tree.


def using_adjacency_list(my_list, nodes_names) :
    visited_list = []
    queue = []
    front = rear = -1 

    for key in my_list.keys() :
        adjacent_nodes = my_list[key]      

        # It will check if node is not present in visited list then it will add both in visited list and queue     
        if key not in visited_list :
            if len(visited_list) == 0 : 
                visited_list.append(key)
                queue.append(key)
                front += 1
                rear += 1
            else :
                visited_list.append(key)
                queue.append(key)
                rear += 1

        # It will check if adjacent nodes of the current key is not present in visited list then it will add both in visited list and queue
        for i in range(len(adjacent_nodes)) :
            if adjacent_nodes[i] not in visited_list :
                    visited_list.append(adjacent_nodes[i])
                    queue.append(adjacent_nodes[i])
                    rear += 1
        
        # If queue turns empty then it will check if visited list and nodes names list are equal then the graph is connected otherwise it is not
        queue.pop(front)
        if (not queue) :
            if visited_list == nodes_names :
                    return True 
            else :
                    return False 
            
    # After iterating on all keys of adjacency list, queue has some element then remove those element and check if visited list = nodes name list then graph is connected    while queue :
    else :
        if visited_list == nodes_names :
                return True 
        else :
                return False 
        

def check_tree(my_list , nodes_names) :

    # This for loop will count the number of nodes in the given graph
    count_nodes = 0
    for key in my_list.keys() :
        count_nodes += 1

    check_connected = using_adjacency_list(my_list , nodes_names)
    my_edge_list = adjacencyMatrix_to_edgeList(adjacencyList_to_adjacencyMatrix(my_list))

    # This for loop will count the number edges in the given graph 
    count_edges = 0 
    for i in range(len(my_edge_list)) :
        count_edges += 1
    
    if check_connected == True and count_edges == (count_nodes - 1) :
        return True
    return False

```

```python
# Q9. Given a tree, find the number of leaf nodes.

def count_leaf_nodes(given_tree) :
    count = 0
    # This for loop will iterate over all keys of given tree and if value of any key is an empty list then it is a leaf node
    for key in given_tree.keys() :
        if given_tree[key] == [] :
            count += 1
    return count
```

```python
# Q10. Given a tree, check if it's a binary tree.

def check_binary_tree(given_tree) :
    # This for loop is iterating over all keys of given tree and check if len of value of all keys should be atmost 2
    for key in given_tree.keys() :
        if len(given_tree[key]) > 2 :
            return False
    return True

```

```python
# Q11. Given a tree and a node, find its height.

def find_nodeHeight(given_node , given_tree) :
    # Base Condition : if the given node is a root node then height = 0.
    if given_tree[given_node] == [] :
        height = 0
    else :
        # Otherwise, whatever is the maximum of the heights of that node's child node + 1 is the height of that node. (calculating using recursion) 
        height = 1 + max([find_nodeHeight(i) for i in given_tree[given_node]])              
    return height

```

```python
# Q12. Given a tree and a node, find its depth.

def find_nodeDepth(given_node , given_tree ) :

    keys_list = list(given_tree.keys())
    root_node = keys_list[0]

    # This for loop will find the parent node of the given node
    for key in given_tree.keys() :
        if given_node in given_tree[key] :
            parent_node = key

    # Base Condition : if given node is root node then its depth = 0.
    if given_node == root_node : 
        depth = 0
    else :
        # Otherwise, whatever is the depth of that node's parent node + 1 is the depth of that node. (calculating using recursion) 
        depth = 1 + find_nodeDepth(parent_node)
    return depth

```
